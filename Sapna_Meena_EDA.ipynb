Kernel Density Plots

```python
datasets = [df1, df2, df3]
for i, df in enumerate(datasets, 1):
    numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
    plt.figure(figsize=(14, len(numerical_columns) * 3))
    for idx, feature in enumerate(numerical_columns, 1):
        plt.subplot(len(numerical_columns), 2, idx)
        sns.histplot(df[feature], kde=True)
        plt.title(f"Dataset {i} - {feature} | Skewness: {round(df[feature].skew(), 2)}")
    plt.tight_layout()
    plt.show()
```

Merging the datasets

```python
df = pd.merge(df3, df1, on="CustomerID", how="left")
df = pd.merge(df, df2, on="ProductID", how="left")
```

```python
Ensure the data types are appropriate
```

```python
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"], errors="coerce")
df["SignupDate"] = pd.to_datetime(df["SignupDate"], errors="coerce")
```

 Drop rows with missing TransactionDate

```python
df = df.dropna(subset=["TransactionDate"])

print("Columns in the merged dataframe:")
print(df.columns, "\n")
```

Analysis

Revenue vs Region

```python
revenue_by_region = df.groupby("Region")["TotalValue"].sum().sort_values(ascending=False)
print(revenue_by_region)
plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_by_region.index, y=revenue_by_region.values, palette="viridis")
plt.title("Total Revenue by Region")
plt.ylabel("Revenue")
plt.xlabel("Region")
plt.show()
```

```python
Revenue vs Product Category
```

```python
revenue_by_category = df.groupby("Category")["TotalValue"].sum().sort_values(ascending=False)
print(revenue_by_category)
plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_by_category.index, y=revenue_by_category.values, palette="muted")
plt.title("Total Revenue by Product Category")
plt.ylabel("Revenue")
plt.xlabel("Category")
plt.show()
```

```python
Region vs Category Analysis
```

```python
region_category = df.groupby(["Region", "Category"]).size().unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(region_category, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Region vs Product Category")
plt.ylabel("Region")
plt.xlabel("Category")
plt.show()
```

```python
Multivarient Analysis
```

```python
Correlation heatmap of numeric variables
```

```python
numeric_columns = [col for col in ["Quantity", "TotalValue", "Price"] if col in df.columns]
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Revenue over Time/Monthly Revenue Analysis

```python
df["TransactionMonth"] = df["TransactionDate"].dt.to_period("M")
monthly_revenue = df.groupby("TransactionMonth")["TotalValue"].sum()
plt.figure(figsize=(14, 8))
monthly_revenue.plot(marker="o", color="b")
plt.title("Monthly Revenue Trend")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.grid()
plt.show()
```
